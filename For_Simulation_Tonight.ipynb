## Types of Simulations

In the study and application of simulations in various fields, several distinct types are commonly used. Each type of simulation has unique characteristics and is suited for specific scenarios. This document provides an overview of several major types of simulations:

- **Deterministic Simulations**: These simulations have no random components. Given a particular set of initial conditions, the output will always be the same.

- **System Dynamics Simulations**: Focuses on the behavior of complex systems over time using stocks, flows, feedback loops, and time delays.

- **Discrete Event Simulations (DES)**: Models the operation of a system as a sequence of events in time. Each event occurs at a particular instant and marks a change of state in the system.

- **Continuous Simulations**: Involves continuous processes, typically described by differential equations. The state of the system changes continuously over time.

```python
# Deterministic simulation of a simple linear growth model
initial_population = 1000
growth_rate = 0.1  # 10% annual growth
years = 10

def linear_growth(initial, rate, time):
    return initial * (1 + rate) ** time

population = linear_growth(initial_population, growth_rate, years)
print(f"Population after {years} years: {population}")
```

# System Dynamics

```python
#### System Dynamics Simulation Example
import numpy as np
import matplotlib.pyplot as plt

# System dynamics model of population growth with births and deaths
time = np.linspace(0, 100, 1000)
birth_rate = 0.02
death_rate = 0.01
initial_population = 1000

def population_growth(t, birth_rate, death_rate, initial_population):
    return initial_population * np.exp((birth_rate - death_rate) * t)

population = population_growth(time, birth_rate, death_rate, initial_population)

plt.plot(time, population)
plt.xlabel('Time (years)')
plt.ylabel('Population')
plt.title('Population Growth Over Time')
plt.show()
```

# Discrete Event Simulation

```python
#### Discrete Event Simulation Example
#!pip install simpy
import simpy

def car(env):
    while True:
        print(f'Car starts parking at {env.now}')
        yield env.timeout(5)
        print(f'Car starts driving at {env.now}')
        yield env.timeout(2)

env = simpy.Environment()
env.process(car(env))
env.run(until=20)
```

# Continuous Simulation

```python
#### Continuous Simulation Example
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Continuous simulation using differential equations (e.g., logistic growth)
def logistic_growth(y, t, r, K):
    dydt = r * y * (1 - y / K)
    return dydt

y0 = 10
t = np.linspace(0, 10, 100)
r = 0.3
K = 50

solution = odeint(logistic_growth, y0, t, args=(r, K))

plt.plot(t, solution)
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Logistic Growth Over Time')
plt.show()
```

# Population Growth Models

## Constant Growth

```python
# Constant growth model
initial_population = 8720000000.0  # World population in 2023
annual_increase = 1e8  # Assume a constant annual increase of 100 million

def constant_growth(initial, increase, years):
    return initial + increase * years

years = 50
future_population = constant_growth(initial_population, annual_increase, years)
print(f"World population after {years} years with constant growth: {future_population}")
```

# Proportional Growth

```python
annual_growth_rate = 0.01  # Assume 1% annual growth

def proportional_growth(initial, rate, years):
    return initial * (1 + rate) ** years

future_population = proportional_growth(initial_population, annual_growth_rate, years)
print(f"World population after {years} years with proportional growth: {future_population}")
```

# Quadratic Growth

```python
quadratic_coefficient = 1e6  # Quadratic term for growth

def quadratic_growth(initial, coefficient, years):
    return initial + coefficient * years ** 2

future_population = quadratic_growth(initial_population, quadratic_coefficient, years)
print(f"World population after {years} years with quadratic growth: {future_population}")
```

# System Dynamics Model

```python
# System dynamics model incorporating births and deaths
birth_rate = 0.02
death_rate = 0.01

def system_dynamics_population(t, birth_rate, death_rate, initial_population):
    return initial_population * np.exp((birth_rate - death_rate) * t)

time = np.linspace(0, 50, 500)
population = system_dynamics_population(time, birth_rate, death_rate, initial_population)

plt.plot(time, population)
plt.xlabel('Time (years)')
plt.ylabel('Population')
plt.title('System Dynamics Model of Population Growth')
plt.show()
```

# Projections
We can use both models to project population growth for the next 50 years and compare the results.

```python
# Comparing deterministic and system dynamics models
years = 50

# Deterministic projections
constant_population = constant_growth(initial_population, annual_increase, years)
proportional_population = proportional_growth(initial_population, annual_growth_rate, years)
quadratic_population = quadratic_growth(initial_population, quadratic_coefficient, years)

# System dynamics projection
system_population = system_dynamics_population(years, birth_rate, death_rate, initial_population)

print(f"Constant Growth Model: {constant_population}")
print(f"Proportional Growth Model: {proportional_population}")
print(f"Quadratic Growth Model: {quadratic_population}")
print(f"System Dynamics Model: {system_population}")
```

# Lanchester's Squared Law & Dupuy's Attrition Model

Lanchester's squared law is used to model the dynamics of combat scenarios. It can be useful in military strategy and operational research.

```python
# Lanchester's squared law
import numpy as np
import matplotlib.pyplot as plt

def lanchester_squared(A0, B0, alpha, beta, t):
    A = A0 - beta * t
    B = B0 - alpha * t
    return A, B

A0 = 1000  # Initial size of force A
B0 = 800   # Initial size of force B
alpha = 0.5  # Effectiveness of A's firepower
beta = 0.3   # Effectiveness of B's firepower
time = np.linspace(0, 100, 500)

A, B = lanchester_squared(A0, B0, alpha, beta, time)

plt.plot(time, A, label='Force A')
plt.plot(time, B, label='Force B')
plt.xlabel('Time')
plt.ylabel('Force Size')
plt.title('Lanchesters Squared Law')
plt.legend()
plt.show()
```

# Dupuy's Attrition Model

```python
# Dupuy's attrition model
def dupuy_attrition(A0, B0, a, b, t):
    A = A0 * np.exp(-a * t)
    B = B0 * np.exp(-b * t)
    return A, B

a = 0.01  # Attrition rate for A
b = 0.015 # Attrition rate for B
time = np.linspace(0, 100, 500)

A, B = dupuy_attrition(A0, B0, a, b, time)

plt.plot(time, A, label='Force A')
plt.plot(time, B, label='Force B')
plt.xlabel('Time')
plt.ylabel('Force Size')
plt.title('Dupuys Attrition Model')
plt.legend()
plt.show()
```
